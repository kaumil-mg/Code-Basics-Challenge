```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Cities = pd.read_csv('Dataset/dim_cities.csv')
Respondent = pd.read_csv('Dataset/dim_repondents.csv')
Survey = pd.read_csv('Dataset/fact_survey_responses.csv')
```

```python
data = pd.merge(Survey,Respondent, on = 'Respondent_ID')
data = pd.merge(data,Cities, on = 'City_ID')
```

```python
# 1. Demographic Insights (examples)
# a. Who prefers energy drink more? (male/female/non-binary?)
gender_preference = data.groupby('Gender')['Consume_frequency'].value_counts().unstack()
gender_preference.plot.barh(stacked=True)
plt.title('Energy Drink Consumption by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion of Responses')
plt.legend(title='Consume Frequency')
plt.show()
```


 As a result we found that males prefer more energy drinks

```python
# b. Which age group prefers energy drinks more?
drinks_more = data[data['Consume_frequency'].isin(['Daily','2-3 times a week'])]
# age_preference = data.groupby("Age")['Consume_frequency'].value_counts().unstack()
age_preference = drinks_more['Age'].value_counts()
age_preference.plot(kind='bar')
plt.title('Consuming energy drink by age')
plt.xlabel('Age')
plt.ylabel('Consumption time')
plt.legend(title ='values for consumption')
plt.show()
```

19-45-age group prefers energy drinks more?

```python
# c. Which type of marketing reaches the most Youth (15-30)?
youth_data = data[data['Age'].isin(['15-18','19-30'])]
marketing_channel = youth_data['Marketing_channels'].value_counts(normalize=True)
marketing_channel.plot(kind = 'bar')
plt.title('Marketing channel trend')
plt.xlabel('Channels')
plt.ylabel('reached')
plt.legend(title='Teams of marketing ')
plt.show()
```

Online ads, TV Commercials type of marketing reaches the most Youth (15-30)

```python
# 2. Consumer Preferences:
# a. What are the preferred ingredients of energy drinks among respondents?

Ingredients = data['Ingredients_expected'].value_counts()
Ingredients.plot(kind='pie',stacked=True)
plt.title('Gender by Ingredient Expected')
# plt.xlabel('Genders')
# plt.ylabel('no. of ingredient expected')
# plt.legend(title='Ingredients')


plt.show()
```

Caffeine, Vitamins are the preferred ingredients of energy drinks among respondents

```python
# Self quetioning at what gender does codesx is being tried
CodeX_used = data[data['Tried_before'].isin(['Yes'])]
CodeX_Gender = CodeX_used['Age'].value_counts()
CodeX_Gender.plot.bar()
plt.title('Age range that tried CodeX')
plt.xlabel('Age Renage')
plt.ylabel('No of person tried CodeX')
plt.legend(title='No. of person tried')
plt.show()
```

```python
# b. What packaging preferences do respondents have for energy drinks?
packaging = CodeX_used['Packaging_preference'].value_counts()
packaging.plot.barh()
plt.title('packaging that prefered by no. CodeX user')
plt.xlabel('CodeX person Prefered in Packaging')
plt.ylabel('Prefered Packaging')
plt.show()
```

Compact and Portable cans, Innovativebottle design packaging preferences do respondents have for energy drinks

```python
# 3. Competition Analysis:
# a. Who are the current market leaders?
# Market_Leaders = data.groupby('Current_brands')['Consume_frequency'].value_counts().unstack()
Market_Leaders = data['Current_brands'].value_counts()
Market_Leaders.plot(kind='pie', figsize=(10,5))
plt.autoscale()
plt.title('Person who drinks a lot')
plt.xlabel('Brands')
# plt.ylabel('No. who drinks a lot')
# plt.legend(title='No. who drinks a lot')
plt.show()
```

Cola-Coka, Bepsi, Gangster are the current market leaders

```python
# b. What are the primary reasons consumers prefer those brands over ours?
prefered_drinking = data['Reasons_for_choosing_brands'].value_counts()
prefered_drinking.plot(kind='pie', stacked=True)
plt.xlabel=('Counts of reason Reasons to choose that brand')
plt.show()
```

Brand reputation, Taste/flavor and Availability are the primary reasons consumers prefer those brands over ours?

```python
# 4. Marketing Channels and Brand Awareness:
# a. Which marketing channel can be used to reach more customers?

marketing_channel.plot.bar()
plt.show()
```

```python
# b. How effective are different marketing strategies and channels in reaching our customers?
# This would involve comparing the responses for brand awareness with the marketing channels used
marketing_effectiveness = data.groupby('Marketing_channels')['Heard_before'].value_counts().unstack().fillna(0)

# Plot the data
marketing_effectiveness.plot.bar()
plt.title('Effectiveness of Marketing Channels in Brand Awareness')
# plt.xlabel('Marketing Channel')
plt.ylabel('Proportion of Responses')
plt.legend(title='Heard About Brand Before')
plt.show()
```

Online ads, TV commercials are different marketing strategies and channels in reaching our customers
