```python
def FunctionAnova(inpData, TargetVariable,ContinuousPridectorList):
    from scipy.stats import f_oneway

    SelectedPredictor =[]

    print('### ANOVA Results ###')
    for predictor in ContinuousPridectorList :
        CategoryGroupLists = inpData.groupby(TargetVariable)[predictor].apply(list)
        AnovaResults = f_oneway(*CategoryGroupLists)

        if(AnovaResults[1]<0.05):
            print(predictor, ' is correlated with ', TargetVariable, '|P-value: ', AnovaResults[1])
            SelectedPredictor.append(predictor)

        else : 
            print(predictor,' is  NOT correlated with ', TargetVariable,' |P-vales: ',AnovaResults[1])

    return(SelectedPredictor)
```

```python
FunctionAnova(ipl,'dream11_score',continuous)
```

```python
features = ['bat_inning',
 'ball_inning',
 'ball_faced',
 'run_scored',
 '4s',
 '6s',
 '50s',
 '100s',
 'ball_delivered',
 'run_given',
 'wicket',
 'catch',
 'stump',]

target_var = 'dream11_score'
```

## ML Modeling

```python
x = ipl[features]
y =ipl[target_var].values
y = y.astype('int')
```

```python
from sklearn.preprocessing import StandardScaler

PredictorScaler= StandardScaler()
PredictorScalerFit = PredictorScaler.fit(x)

x=PredictorScalerFit.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
```

```python
regrassor = [AdaBoostRegressor(), RandomForestRegressor(),LinearRegression(),DecisionTreeRegressor()]

name =[]
train_acc=[]
test_acc=[]
models=[]
rmse=[]
i=0
for regression in regrassor:
    regression.fit(x_train,y_train)
    name.append(type(regression).__name__)
    train_acc.append(regression.score(x_train,y_train))
    test_acc.append(regression.score(x_test,y_test))
    models.append(regression)
    rmse.append(np.sqrt(mean_squared_error(regression.predict(x_test),y_test)))
```

```python
df_score=pd.DataFrame(list(zip(name,train_acc,test_acc,models,rmse)),columns=['name','train_acc','test_acc','models','rmse'])
df_score.set_index('name',inplace=True)
df_score.sort_values('test_acc',inplace=True)
df_score
```

```python
model = df_score.loc['LinearRegression','models']
prediction = model.predict(x_test)
```

```python
print(prediction)
```

```python
%matplotlib inline
features_importance = pd.Series(model.coef_,index=features)
features_importance.nlargest(10).plot(kind='barh')
```
